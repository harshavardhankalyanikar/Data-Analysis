```python
# Replace the nan values in LoanAmount column with median value
df.loc[df['LoanAmount'].isnull()==True,'LoanAmount']=np.median(df.LoanAmount.dropna(axis=0))
```

```python
# replace the nan values in Loan_Amount_Term with the median value
df.loc[df['Loan_Amount_Term'].isnull()==True,'Loan_Amount_Term']=np.median(df.Loan_Amount_Term.dropna(axis=0))
```

```python
if df['Married'].isnull().sum()>0:
    df.loc[df['Married'].isnull()==True,'Married']=df['Married'].mode()[0]
```

```python
df.info()
```

```python
if df['Dependents'].isnull().sum()>0:
    df.loc[df['Dependents'].isnull()==True,'Dependents']=df['Dependents'].mode()[0]
```

```python
df.info()
```

```python
if df['Gender'].isnull().sum()>0:
    df.loc[df['Gender'].isnull()==True,'Gender']=df['Gender'].mode()[0] 
```

```python
df.info()
```

```python
if df['Self_Employed'].isnull().sum()>0:
    df.loc[df['Self_Employed'].isnull()==True,'Self_Employed']=df['Self_Employed'].mode()[0]
```

```python

# Loan Status Count Plot
sns.countplot(x='Loan_Status (Approved)', data=df)
plt.title('Loan Approval Status')
plt.show()
```

# Insights

This plot gives us the idea about the acceptance rate of loan from the bank

```python
# Loan Status by Gender
sns.countplot(x='Gender', hue='Loan_Status (Approved)', data=df)
plt.title('Loan Status by Gender')
plt.show()
```

# Insights

It gives us the difference between approval rate between the genders

```python
# Loan Status by Marital Status
sns.countplot(x='Married', hue='Loan_Status (Approved)', data=df)
plt.title('Loan Status by Marital Status')
plt.show()
```

# Insights

This gives us info about loan acceptance by martial status

```python
# Loan Status by Education Level
sns.countplot(x='Education', hue='Loan_Status (Approved)', data=df)
plt.title('Loan Status by Education Level')
plt.show()
```

# Insights

Applicants with higher education may have a higher approval rate due to perceived financial stability

```python
# Applicant and Coapplicant Income Distribution

sns.histplot(df['ApplicantIncome'], bins=30)
plt.title('Applicant Income Distribution')
plt.show()

sns.histplot(df['CoapplicantIncome'], bins=30)
plt.title('Coapplicant Income Distribution')
plt.show()
```

# Insights

Most applicants earn below a certain threshold. Income distribution helps understand who is more likely to get approved.

```python
# loan approval by loan amount

sns.boxplot(x='Loan_Status (Approved)', y='LoanAmount', data=df)
plt.title('Loan Amount vs Approval Status')
plt.show()
```

# Insights

Approved loans have a median loan amount of X, while rejected loans tend to have higher or lower amounts. Helps in understanding bank risk assessment.

```python
# Loan status by Credit History

sns.countplot(x='Credit_History', hue='Loan_Status (Approved)', data=df)
plt.title('Loan Status by Credit History')
plt.show()
```

# Insights

Credit history strongly influences approval. Most approvals happen when credit history is good (1).

```python
# Loan approval by Property Area

sns.countplot(x='Property_Area', hue='Loan_Status (Approved)', data=df)
plt.title('Loan Status by Property Area')
plt.show()
```

# Insights

Applicants from semi-urban areas seem to get more approvals. Helps target marketing or financial products.

# Conclusion

Approval trends: 68% approved, 32% rejected

Income & Education: Higher income & educated applicants more likely to get loans
